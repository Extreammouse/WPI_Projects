```python
# Imports

import io
import time
import torch

from PIL import Image
from diffusers import DiffusionPipeline
from datasets import load_dataset, Dataset, DatasetDict, concatenate_datasets
```

```python
pipeline = DiffusionPipeline.from_pretrained("stable-diffusion-v1-5/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipeline.to("cuda")
```

```python
# Load the existing dataset from Hugging Face (or create a new one)
dataset_name = "DiegoP-S/DatasetSynthesis"
dataset = load_dataset(dataset_name)

```

```python
samples_per_class = 100
batch_size = 20
group = "mbkinaci/fruit-images-for-object-detection"
classes = {"Apple", "Banana", "Beetroot", "Bell Pepper", "Cabbage", "Capsicum", "Carrot", "Cauliflower", "Chilli Pepper", "Corn", "Garlic", "Ginger", "Jalapeno", "Kiwi", "Mango", "Orange", "Paprika", "Peas", "Pineapple", "Pomegranate", "Potato", "Soybean", "Sweet Potato", "Sweetcorn", "Tomato", "Watermelon", "Pear", "Grapes", "Cucumber", "Onion", "Lemon", "Spinach", "Turnip", "Eggplant", "Dog", "Horse", "Elephant", "Butterfly", "Chicken", "Cat", "Cow", "Sheep", "Squirrel", "Dog", "Spider"}

for class_name in classes:
    if class_name.replace(" ", "_") not in dataset.keys():
        split = {class_name: {"image": [], "file_name": []}}
        print(f"Creating {class_name} split")
    else:
        split = {class_name: dataset[class_name.replace(" ", "_")].to_dict()}
        print(f"Resuming {class_name} split")
    for i in range(samples_per_class):
        file_name = f"{class_name}_generated_image_{i}.png"

        # Check if image already in dataset
        if class_name.replace(" ", "_") in dataset.keys() and file_name in dataset[class_name.replace(" ", "_")]["file_name"]:
            print(f"Skipping {file_name}")
            continue

        # Generate the image using the pipeline
        print(file_name)
        generated_image = pipeline(f"{class_name}").images[0]
        generated_image.save("tmp.png", format='PNG')

        image = Image.open("tmp.png")  # Open the image
        bytes = io.BytesIO()
        image.save(bytes, format="PNG")

        # Convert the image to a binary format (e.g., PNG)
        img_byte_arr = io.BytesIO()
        generated_image.save(img_byte_arr, format='PNG')
        img_byte_arr = img_byte_arr.getvalue()

        # You can now upload the binary data to Hugging Face Dataset
        split[class_name]["file_name"].append(file_name),
        split[class_name]["image"].append(bytes.getvalue()),

        if i % batch_size == 1:
            # Create a new dataset with the updated data
            dataset[class_name.replace(" ", "_")] = Dataset.from_dict(split[class_name])

            # Push the updated dataset to Hugging Face
            dataset.push_to_hub(dataset_name)
    # Create a new dataset with the updated data
    dataset[class_name.replace(" ", "_")] = Dataset.from_dict(split[class_name])

    # Push the updated dataset to Hugging Face
    dataset.push_to_hub(dataset_name)
```
